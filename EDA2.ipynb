```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
df= pd.read_csv('data/data_train.csv', delimiter='\t', quotechar='"')
```

```python
df.
```

## Nombre de fille et de garçon

```python
df['sex'].value_counts(normalize=True)
```

Il y a 59% de fille et 41% de garçon

## Nombre d'élève de chaque école

```python
df['school'].value_counts(normalize=True)
```

Il y a 67% d'élève de l'école GP et 33% d'élève de l'école MS

## Moyenne des notes

```python
# plot la moyenne des notes en fonction du sexe et de l'école
df.groupby('sex')['G3'].mean().plot(kind='bar')
plt.ylabel('Moyenne des notes')
plt.title('Moyenne des notes en fonction du sexe et de l\'école')
plt.show()

# plot la moyenne des notes en fonction du sexe et de l'école
df.groupby('school')['G3'].mean().plot(kind='bar')
plt.ylabel('Moyenne des notes')
plt.title('Moyenne des notes en fonction du sexe et de l\'école')
plt.show()
```

Les filles ont généralement de meilleures notes que les garçons. Les élèves de l'école GP ont généralement de meilleures notes que les élèves de l'école MS

## Analyse  sur la consommation d'alcool

```python

# Supposons que ton DataFrame s'appelle df et qu'il contient les colonnes 'Dalc' et 'G3'

plt.figure(figsize=(10, 6))
# Graphique de dispersion avec ligne de régression
sns.regplot(data=df, x='Dalc', y='G3', scatter_kws={'s': 50}, line_kws={"color": "red"})
plt.title('Relation entre la consommation d\'alcool pendant la semaine (Dalc) et les performances scolaires (G3)')
plt.xlabel('Consommation d\'alcool en semaine (Dalc)')
plt.ylabel('Note finale G3')
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
# Graphique de dispersion avec ligne de régression
sns.regplot(data=df, x='Walc', y='G3', scatter_kws={'s': 50}, line_kws={"color": "red"})
plt.title('Relation entre la consommation d\'alcool pendant la semaine (Dalc) et les performances scolaires (G3)')
plt.xlabel('Consommation d\'alcool en semaine (Dalc)')
plt.ylabel('Note finale G3')
plt.grid(True)
plt.show()
```

on peux voir que la consommation d'alcool pendant la semaine et le week-end n'ont une légére influence sur les notes des élèves

# analyse des activités extra-scolaires en fonction des notes

```python


plt.figure(figsize=(10, 6))
# Graphique de dispersion avec ligne de régression
sns.regplot(data=df, x='freetime', y='G3', scatter_kws={'s': 50}, line_kws={"color": "red"})
plt.title('temps libre après l\'école (freetime) et les performances scolaires (G3)')
plt.xlabel('temps libre après l\'école (freetime)')
plt.ylabel('Note finale G3')
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
# Graphique de dispersion avec ligne de régression
sns.regplot(data=df, x='goout', y='G3', scatter_kws={'s': 50}, line_kws={"color": "red"})
plt.title('temps passé avec les amis (goout) et les performances scolaires (G3)')
plt.xlabel('temps passé avec les amis (goout)')
plt.ylabel('Note finale G3')
plt.grid(True)
plt.show()
# Analyse de l'influence de freetime et goout sur G3
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='freetime', y='G3')
plt.title('Influence du temps libre après l\'école (freetime) sur les performances scolaires (G3)')
plt.xlabel('Temps libre après l\'école (freetime)')
plt.ylabel('Note finale G3')
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='goout', y='G3')
plt.title('Influence du temps passé avec les amis (goout) sur les performances scolaires (G3)')
plt.xlabel('Temps passé avec les amis (goout)')
plt.ylabel('Note finale G3')
plt.show()
```

```python
# on peux voir que le temps libre et les sorties avec les amis ont une influence sur les notes des élèves en général les éléves qui sont modérer et qui sont à deux ont des meilleurs notes. puis on peux aperçevoir que les éléves qui sortent beaucoup ont des notes plus basse que les autres.une légére baisse des résulats aprés.
```

```python
# Sélection des variables familiales et des performances scolaires
family_vars = ['Pstatus', 'famsize', 'famrel', 'famsup', 'G3']

# Conversion des variables catégorielles en numériques si nécessaire
df['Pstatus'] = df['Pstatus'].map({'T': 1, 'A': 0})
df['famsize'] = df['famsize'].map({'GT3': 1, 'LE3': 0})
df['famsup'] = df['famsup'].map({'yes': 1, 'no': 0})

# Matrice de corrélation
plt.figure(figsize=(8, 6))
sns.heatmap(df[family_vars].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Matrice de corrélation entre les variables familiales et les performances scolaires')
plt.show()
```
